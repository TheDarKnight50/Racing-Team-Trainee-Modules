```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = np.arange(2020, 2025)
india_birth_rate = np.random.uniform(20, 30, size=5)
nepal_birth_rate = np.random.uniform(15, 25, size=5)
bhutan_birth_rate = np.random.uniform(10, 20, size=5)

plt.figure(figsize=(10, 6))

plt.plot(years, india_birth_rate, marker='o', color='blue', label='India')
plt.plot(years, nepal_birth_rate, marker='s', color='green', label='Nepal')
plt.plot(years, bhutan_birth_rate, marker='^', color='red', label='Bhutan')

plt.title('Birth Rate Comparison')
plt.xlabel('Year')
plt.ylabel('Birth Rate (per 1000 people)')
plt.xticks(years)
plt.legend()

plt.grid(True)
plt.tight_layout()

plt.show()
```
