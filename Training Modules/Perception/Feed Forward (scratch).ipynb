```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
```

```python
(X_train, Y_train), (X_test, Y_test) = tf.keras.datasets.mnist.load_data()
X_train = X_train.reshape(-1, 784).astype('float32') / 255.0
X_test = X_test.reshape(-1, 784).astype('float32') / 255.0
Y_train = Y_train.astype('int32')
Y_test = Y_test.astype('int32')

validation_split = 0.1
split_index = 5000
X_val = X_train[split_index:6000]
Y_val = Y_train[split_index:6000]
X_train = X_train[:split_index]
Y_train = Y_train[:split_index]
```

```python
i=784
h1=16
h2=16
o=10
lr=0.2
epochs=10

W1=np.random.randn(h1,i)
b1=np.zeros((h1,1))
W2=np.random.randn(h2,h1)
b2=np.zeros((h2,1))
W3=np.random.randn(o,h2)
b3=np.zeros((o,1))

def sigmoid(z) :
    return (1/(1+np.exp(-z)))
def sigmoidprime(z) :
    return sigmoid(z)*(1-sigmoid(z))

def one_hot(y,bs) :
    oh=np.zeros((bs,10))
    oh[np.arange(bs),y]=1
    return oh

losslist=[]
valacclist=[]

#training loop
for ep in range(epochs):
    for i in range(5000):

        X_batch = X_train[i]
        Y_batch = Y_train[i]

        #feed forward
        z1=np.dot(X_batch,W1.T)+b1.T
        a1=sigmoid(z1)
        z2=np.dot(a1,W2.T)+b2.T
        a2=sigmoid(z2)
        z3=np.dot(a2,W3.T)+b3.T
        y_pred=sigmoid(z3)

        Y_batch_enc=one_hot(Y_batch,1)
        loss=np.mean((Y_batch_enc - y_pred) ** 2)
        losslist.append(loss)

        xb=X_batch.reshape(1,784)

        da3=-(Y_batch_enc - y_pred)*2
        dW3=np.dot(np.diag((np.array(sigmoidprime(z3)))[0,:]),np.dot(da3.T,a2))
        db3=np.dot(np.diag((np.array(sigmoidprime(z3)))[0,:]),da3.T)

        da2=np.dot(W3.T,np.dot(np.diag((np.array(sigmoidprime(z3)))[0,:]),da3.T))
        dW2=np.dot(np.diag((np.array(sigmoidprime(z2)))[0,:]),np.dot(da2,a1))
        db2=np.dot(np.diag((np.array(sigmoidprime(z2)))[0,:]),da2)

        da1=np.dot(W2.T,np.dot(np.diag((np.array(sigmoidprime(z2)))[0,:]),da2))
        dW1=np.dot(np.diag((np.array(sigmoidprime(z1)))[0,:]),np.dot(da1,xb))
        db1=np.dot(np.diag((np.array(sigmoidprime(z1)))[0,:]),da1)

        W1-=lr*dW1
        b1-=lr*db1
        W2-=lr*dW2
        b2-=lr*db2
        W3-=lr*dW3
        b3-=lr*db3

    z1=np.dot(X_val,W1.T)+np.dot(np.ones((1000,1)),b1.T)
    a1=sigmoid(z1)
    z2=np.dot(a1,W2.T)+np.dot(np.ones((1000,1)),b2.T)
    a2=sigmoid(z2)
    z3=np.dot(a2,W3.T)+np.dot(np.ones((1000,1)),b3.T)
    y_pred_val=sigmoid(z3)
    valacc=np.mean(np.argmax(y_pred_val,axis=1)==Y_val)
    print(f"validation accuracy epoch {ep+1} = {valacc}")
    valacclist.append(valacc)
```

```python
def classify(image):
        z1=np.dot(image.reshape((1,784)),W1.T)+b1.T
        a1=sigmoid(z1)
        z2=np.dot(a1,W2.T)+b2.T
        a2=sigmoid(z2)
        z3=np.dot(a2,W3.T)+b3.T
        y_pred=sigmoid(z3)
        return np.argmax(y_pred,axis=1)
```
